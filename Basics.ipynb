NumPy (Numerical Python)

```python
import numpy as np

#3x3 random matrix
arr = np.random.randint(0,10,(3,3)) 
print("Matrix:\n" ,arr)

#Row sum
print("Row sum:\n", arr.sum(axis=1))

#Column max
print("Max column:\n", arr.max(axis=0))

#Sum of all elements
print("Total Sum:\n", arr.sum(axis=None))

#Transpose of arr
print("Transpose:\n", arr.T)
```

Data Preprocessing

```python
scores = np.array([65,70,90,95,80])
print("Scores:\n", scores)

#Normalization (x-min)/(max-min)
norm_scores = (scores - scores.min()) / (scores.max() - scores.min())
print("Normalized:\n", norm_scores )
```

```python
#Weight matrix (3x2)
W = np.array([[2,-1],[0,3],[1,4]])

#Input Vector (2x1)
x = np.array([[5],[2]])

#Matrix multiplication
output = W.dot(x)
print("Dot Product: \n", output)
```

```python
#Dataset
X = np.array([1,2,3,4,5])
#Relation
y = 2*X

#Initialize weight
w = np.random.randn()

#Learning rate
lr = 0.01

#Gradient Descent
for epoch in range(100):
    y_pred = w*X
    loss = ((y - y_pred)**2).mean()
    grad = -2*(X*(y-y_pred)).mean()
    w -= lr * grad 
    
print("Learned Weight:\n",w)
```

Pandas


```python
import pandas as pd
#Series (1D)
s = pd.Series([10,20,30], index=['a','b','c'])
print("Series:\n",s)

#Dataframe(2D)
data = {
    'Name':['A','B','C'],
    'Age':[20,25,30],
    'Score': [90,80,70]
}
df = pd.DataFrame(data)
print("Dataframe:\n",df)
```

```python
import pandas as pd

data = {
    'Name' : ['A','B','C','D','E'],
    'Age' : [20 , 25, None, 30, 28],
    'Gender' : ['M', 'F', 'F', 'M', 'F'],
    'Score' : [90, 85, None, 70, 60]
}
df = pd.DataFrame(data)
# printing data
print("Date: \n", df)

#Missing values count
print("Missing values count:\n", df.isnull().sum())

#Fill missing Age with mean
df.fillna({'Age' : df['Age'].mean()}, inplace = True)
print("Date: \n", df)

#Fill missing score with median
df.fillna({'Score' :  df['Score'].median()}, inplace=True)
print("Date: \n", df)

#Filter: female with Score > 70
print("Filtered value:\n",df[(df["Gender"] == 'F') & (df['Score']>70)])

#Add AgeGroup column
df['AgeGroup'] = df['Age'].apply(lambda x : 'Young' if x<25 else 'Old')
print(df)
```

Multiplots

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("Sine")
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("Cosine")

plt.show()
```

Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt

#Built-in Datasets
tips = sns.load_dataset("tips") #Restaurant tips dataset
iris = sns.load_dataset("iris") #Famous flower dataset

#Styling
sns.set_style("whitegrid")

plt.figure(figsize=(12, 28)) 

#1: Distribution Plots
plt.subplot(5,1,1)
sns.histplot(tips["total_bill"], kde=True)
plt.show()

#2: Scatter plot
plt.subplot(5,1,2)
sns.scatterplot(x="total_bill", y = "tip", data=tips, hue="sex")
plt.show()

#3: Box Plot
plt.subplot(5,1,3)
sns.boxplot(x="day", y = "total_bill", data=tips)
plt.show()

#4: Pair Plot
plt.subplot(5,1,4)
sns.pairplot(iris, hue="species")
plt.show()

#5: Heatmap
plt.subplot(5,1,5)
corr = tips.corr(numeric_only = True)
sns.heatmap(corr, annot= True,cmap = "coolwarm")
plt.show()
```
